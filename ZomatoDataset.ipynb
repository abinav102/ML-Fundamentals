```python
import pandas as pd
zomato = pd.read_csv('zomato.csv',encoding = 'latin-1')
zomato.head()
```

```python
zomato.info()
```

```python
zomato.isnull().sum()
```

9 missing values for cuisine , which has a datatype of 'object'----> could be string values


Dropping the missing rows using .dropna() function


```python
zomato.dropna(subset=['Cuisines'],inplace = True)  #subset=['Cuisines']-->look for nulls only in that column
zomato.info()                                                 #inplace = True--> modify the dataframe directly
```

Changing Yes/No values into binary

```python
category = ['Has Table booking','Has Online delivery','Is delivering now','Switch to order menu']
for cols in category:
    zomato[cols] = zomato[cols].map({'Yes':1,'No':0})
zomato.head()
```

PRINTING THE TOP 5 RESTAURANTS BASED ON **RATING**

```python
grouped_by_locality = zomato.groupby('Locality')
avg_rating = grouped_by_locality['Aggregate rating'].mean()
top_5 = avg_rating.sort_values(ascending = False)
top_5.head()
```

```python
restaurant_count = zomato.groupby('Locality').size()
locality_stats = pd.DataFrame({'Average Rating ' : avg_rating , 'Restaurant count': restaurant_count})
locality_stats.head()
```

```python
reliable_localities = locality_stats[locality_stats['Restaurant count']>30]
top_10 = reliable_localities.sort_values(by = 'Average Rating ',ascending = False)
top_10.head()
```

# **PLOTTING** **A** **GRAPH**

```python
rating_counts = zomato['Rating text'].value_counts()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (10,6))  #set the figure size---> 10 inches wide,6 inches tall
sns.barplot(x = rating_counts.index, y = rating_counts.values)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
```

Dealing with the 'not rated' restaurants

```python
not_rated = zomato[zomato['Rating text']=='Not rated']
not_rated['Votes'].describe()
```

```python
top_cuisines = zomato['Cuisines'].value_counts().head(10)
top_cuisines.head()





































```

split function is applied only if x belongs to str

```python

top_10_cuisines = zomato['Cuisines'].explode().value_counts().head(10)
```

```python

plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_cuisines.index, y=top_10_cuisines.values)
plt.title('Top 10 Most Common Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Average Cost for two',y='Aggregate rating',data=zomato)
plt.title("Cost vs Rating Analysis")
plt.show()
```
